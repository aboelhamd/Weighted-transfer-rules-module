```python
import os

models_path='sklearn-models'
model_name='LinearSVM'
output_path='output.csv'
dataset_path='/home/aboelhamd/Downloads/sklearn-svm/sklearn-nobad'
files = {}
# r=root, d=directories, f=files
for r, d, f in os.walk(dataset_path):
    for file in f:
        files[file]=os.path.join(r, file)

# os.mkdir('models')
```

```python
### import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
from matplotlib.colors import ListedColormap
from sklearn.model_selection import train_test_split
from sklearn.svm import SVC
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier, AdaBoostClassifier
from sklearn.naive_bayes import GaussianNB
from sklearn.externals import joblib
from sklearn.preprocessing import OrdinalEncoder


for file in files:
    # These are the classifiers that permit training data with sample weights!
#     names = ["NaiveBayes", "LinearSVM", "RBFSVM", "DecisionTree",
#          "RandomForest", "AdaBoost"]
    
    print("file name :", file)
    data = pd.read_csv(files[file], delimiter=r"\s+", header=None).dropna()
    
    # words (features) encoding
    # load the encoder
    enc = joblib.load(models_path+'/'+'encoder'+'-'+file[:-4])
    # remove records with unseen word, will return always 0 for that record
    # this will be solved later
    unseen = []
    for i in range(len(data.values)) :
        for j in range(len(data.values[i])) :
            if data.values[i][j] not in enc.categories_[j] :
                unseen.append(data.values[i])
    
    seen = [x for x in data.values if x not in unseen]

    samples = enc.transform(seen)
    
#     print("file name :", file)
#     print("Rules(classes) number :",target.nunique())
    print("Words(features) number :",samples.shape[1])
    print("Records number :",samples.shape[0], end = '')
#     display(data.iloc[:target.nunique(),:])
    
#     display(features, target, weights)
    
    # prediction by using svm
#     print("model :", name, ",", end = '')
    name = 'LinearSVM'
    modelname = models_path+'/'+name+'-'+file[:-4]+'.model'
    loaded_model = joblib.load(modelname)
    rules = loaded_model.predict(samples)
    # insert zero in unseen data
    for i in range(len(data.values)) :
        if data.value[i] in unseen :
            rules.insert(i,0)

    # write results in file
    output = open(output_path, 'w+')
    for rule in rules :
        output.write(rule+"\n")
    
    print("----------------------------------------------\n")
```
